```python
import pandas as pd 
import numpy as np 
from sklearn.impute import SimpleImputer #simpleimputer is basically used to handle missing values.
```

```python
df = pd.read_csv('Data.csv')
```

```python
df
```

```python
#initialising the imputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

## Removing NaN values from Country column

```python
df.iloc[:,0:1].values
```

```python
imputer.fit(df.iloc[:,0:1].values)
final_values = imputer.transform(df.iloc[:,0:1].values)
```

```python
df.iloc[:,0:1] = final_values
```

```python
df
```

## Removing NaN values from all the colums - 

```python
#calculating and performing the most frequent strategy for the whole dataset and all columns so no need to use iloc - 
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:] = imputer.transform(df.iloc[:,:].values)
```
