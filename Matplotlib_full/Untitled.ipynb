```python
import matplotlib as mp
import numpy as np
import matplotlib.pyplot as plt
```

### Generating Sin and Cos graph 

```python
x = np.linspace(0, np.pi, 50) # this will create 10 numbers with equal variance
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x,sin_y, label='sin') #either we can use the label parameter to set manual labels or use signs like below
plt.plot(x,cos_y, label='cos') 


plt.legend(loc='upper right') #this is essential to show the labels 
# plt.legend(loc='upper left') # this moves the label to the upper left corner 
#here are all the parameters we can put inside loc of plt.legend(loc='')
#upper right/left
#center
#lower right/left
plt.title("Sin Wave")
plt.show()
```

```python
plt.plot(x,sin_y, '*') 
plt.plot(x,cos_y, '+') 
plt.show()
```

#### Using the linestyle command. It gives us 4 different styles for use - 

```python
plt.figure(figsize=(15,5))
plt.plot(x,x+0, linestyle='solid', label='Solid')
plt.plot(x,x+1, linestyle='dashed', label='Dashed') #using x+1 as y will do noting but add one to all the values of x, so the value 1 will become 2 and so on.
plt.plot(x,x+2, linestyle='dashdot', label='DashDot')
plt.plot(x,x+3, linestyle='dotted', label='Dotted')
plt.legend()
```

#### using xlim and ylim to set values of x and y axis - 
so basically it can be used to set custom values for x and y axis rather than making then auto appear

```python
x = np.linspace(0, 4*np.pi, 100)
x_sin = np.sin(x)

plt.xlim([-1,15]) #setting x axis from -1 to 15
plt.ylim([-2,2]) #setting y axis from -2 to 2

# plt.axis('equal')
# plt.axis('tight')

# plt.xlim([5,15]) #maybe you just wanna show values between 5 and 15

plt.plot(x,x_sin)
```

#### Pie Chart - 

##### labels = will be the labels used for your data, 
##### explode = takes the same value as alpha and makes the values that are not 0 stand out.

```python
cars = ['AUDI','MERCEDES','FORD','TESLA','SUZUKI','TOYOTA','TATA']
market_share = [ 50, 55, 20 , 4, 300, 250, 200]
explode = [0,0,0,0,0,0,.1]
alpha = [0,0,0,0,0,0,.1]
fig = plt.figure(figsize=(7,7))
plt.pie(market_share, labels=cars, explode=explode)
```

`Adding Shadow`

```python
cars = ['AUDI','MERCEDES','FORD','TESLA','SUZUKI','TOYOTA','TATA']
market_share = [ 50, 55, 20 , 4, 300, 250, 200]
explode = [0,0,0,0,0,0,.1]
alpha = [0,0,0,0,0,0,.1]
fig = plt.figure(figsize=(7,7))
plt.pie(market_share, labels=cars, explode=explode, shadow=True)
```

`Adding Percentages`

```python
cars = ['AUDI','MERCEDES','FORD','TESLA','SUZUKI','TOYOTA','TATA']
market_share = [ 50, 55, 20 , 4, 300, 250, 200]
explode = [0,0,0,0,0,0,.1]
alpha = [0,0,0,0,0,0,.1]
fig = plt.figure(figsize=(7,7))
plt.pie(market_share, labels=cars, explode=explode, autopct='%1.1f%%')
```

`Adding Wedget`

```python
cars = ['AUDI','MERCEDES','FORD','TESLA','SUZUKI','TOYOTA','TATA']
market_share = [ 50, 55, 20 , 4, 300, 250, 200]
explode = [0,0,0,0,0,0,.1]
alpha = [0,0,0,0,0,0,.1]
fig = plt.figure(figsize=(7,7))
plt.pie(market_share, labels=cars, explode=explode, autopct='%1.1f%%', 
        wedgeprops={'edgecolor':'black', 'linewidth':2, 'antialiased':True})
```

#### Multiple Subplots - 

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title("Sin value")

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title("Cos value")

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title("Tan value")

y = np.random.randn(1000)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title("Hist Plot value")

plt.tight_layout() #to make a little gap between the upper and lower ones.
```

`Example 2`

```python
x = np.linspace(0,10,50)
y = np.sin(x)

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title("Sin wave")
ax2.set_title("Sin wave - square")
ax3.set_title("Sin wave - cube")

plt.tight_layout()
```
