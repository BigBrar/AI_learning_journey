## Following are some methods to handle missin

```python
import pandas as pd
df = pd.read_csv('AB_NYC_2019.csv')
```

## Deleting the column with missing data - 

```python
df1 = df.copy()
df1.drop('last_review', axis=1, inplace=True)
df['last_review']
```

## Deleting row with the missing data - 

```python
df2 = df.copy()
df2.dropna(inplace=True)
print(f'Previous values: {df1.shape}, Values after dropna: {df2.shape}')
```

## Filling the missing values 
Numerical Values (using mean) :

```python
df3 = df.copy()
```

```python
#finding the mean/average value for reviews_per_month - 
mean_value = df3['reviews_per_month'].mean()
float(mean_value)
df3['reviews_per_month'].fillna(mean_value)
df
```

```python
df3.head()
```

## Filling categorical data

```python
#value_counts() calculates the mod by getting the most used value. So in this case we will replace null values with mostly used ones that is mod...
most_repeated_value = df3['last_review'].value_counts().index[0] #selecting the first one.
most_repeated_value
```

```python
df3['last_review'].fillna(most_repeated_value)
```

```python
df3.head()
```

```python
#it's not like we only have to replace the values with mod we can also use something else like a manual value like "Not_reviewed"
df4 = df.copy()
df4['last_review'] = df4['last_review'].fillna('Not_reviewed')
```

```python
df4.head()
```

## Advanced Imputation (Interpolate)- 
Interpolate method can only be used on numerical data and it basically uses the previous and next value of the NaN value to replace the NaN value, so it basically generates the mean of the previous and next value and places it at the place of NaN that will be in between.

```python
df5 = df.copy()
df5['reviews_per_month'] = df5['reviews_per_month'].interpolate()
```
