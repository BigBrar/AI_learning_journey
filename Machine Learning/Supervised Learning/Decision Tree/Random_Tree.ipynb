```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('WineQT.csv')
dataframe.head()
```

```python
dataframe.columns
```

```python
set(dataframe['quality'])
```

Splitting the data

```python
x = dataframe.drop(columns='quality', axis=1)
y = dataframe['quality']
```

```python
x.columns
```

splitting data into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

Data Modeling

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

Model Predictions

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(model, filled=True)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted') # Use weighted average for multiclass classification
recall = recall_score(y_test, y_pred, average='weighted') # Use weighted average for multiclass classification

print(f'Accuracy: {accuracy}')
print(f'F1 Score: {f1}')
print(f'Recall: {recall}')
```

Implementing Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test)
```

```python
y_predict = rf.predict(x_test)
print(y_predict)
```

```python
rf.estimators_
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(rf.estimators_[0], filled=True)
```

HyperParameter Tuning: GridSearchCV & RandomSearchCV

```python
# letting the model experiment with all different parameters to find out the best one
grid_param = {
    'n_estimators': [50, 100,120],
    'criterion': ['gini','entropy'],
    'max_depth': range(5)
    # 'max_depth': range(10),
    # 'min_samples_leaf': range(10)
}

```

```python
from sklearn.model_selection import GridSearchCV
grid_search_rf = GridSearchCV(param_grid=grid_param, cv=10, n_jobs=6, verbose=1, estimator=rf)
grid_search_rf.fit(x_train, y_train)
```

```python
#finding out which are the best parameter for best results according to model -
grid_search_rf.best_params_
```

```python
# now creating a new model with given parameters
rf_new = RandomForestClassifier(n_estimators=120, max_depth=4, criterion='entropy')
rf_new.fit(x_train, y_train)
```

```python
# checking the new scores
rf_new.score(x_test, y_test)
```
