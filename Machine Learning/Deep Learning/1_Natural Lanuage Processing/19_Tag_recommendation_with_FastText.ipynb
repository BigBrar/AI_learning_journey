```python
import pandas as pd
df = pd.read_csv('products.csv')
```

```python
df.head(2)
```

```python
df['features'] = df['product_name'] + ' ' + df['product_tag'] + ' ' + df['brand_tag']
df['tags'] = ',' + df['tags']
df['label'] = ['__label__'.join(tag.split(',')).strip().lower() for tag in df['tags']]
df.head(2)
```

```python
data = [i[-1] + ' ' + i[-2] for i in df.values if pd.notna(i[-1]) and pd.notna(i[-2])]
fd = open('data.txt','w')
for i in data:
  fd.write(i)
  fd.write('\n')

fd.close()
```
