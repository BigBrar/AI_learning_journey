```python
df2 = sns.load_dataset('tips')
df2.head()
```

```python
sns.scatterplot(x='total_bill', y='tip' ,data=df2)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex' ,data=df2)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker' ,data=df2) #added a third category for smoker 
```

## JointPlot -
gives the distribution as well as the scatter plot

```python
sns.jointplot(x='total_bill', y='tip', data=df2)
```

```python
sns.jointplot(x='total_bill', y='tip', hue='sex' ,data=df2)
```

```python
sns.jointplot(x='total_bill', y='tip' ,data=df2, kind='reg')
```

```python
df2.corr(numeric_only=True)
#correlation tells you how all the variables are linearly connceted so if the values of one changes how would it affect the others. So basically it 
# tells how exactly are they correlated.
```
