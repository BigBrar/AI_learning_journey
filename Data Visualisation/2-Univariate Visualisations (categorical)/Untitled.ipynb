```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
```

## Categorical visualisation - 

```python
# sns.countplot(data = df, x = 'neighbourhood_group') # this has same color for all bars in graph.
sns.countplot(data = df, x = 'neighbourhood_group', hue='neighbourhood_group')
```

```python
sns.countplot(data = df, x = 'room_type', hue='room_type')
```

```python
sns.countplot(data = df, x = 'neighbourhood_group', hue='room_type') #display room types in neighbourhood groups
```

## Pie chart - 

```python
# df['room_type'].value_counts().plot(kind = 'pie') without percentage
df['room_type'].value_counts().plot(kind = 'pie', autopct = '%.2f') #with percentage where 2f denotes 2 decimals after '.' 
```

```python
df['neighbourhood_group'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['neighbourhood_group'].value_counts().plot(kind='bar') # we can also create bar graphs without seaborn by just using .plot()
```
