```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.countplot(x='survived', hue='class', data=df)
```

```python
#Crosstabs are a valuable tool for analyzing the relationship between two categorical variables
x = pd.crosstab(df['survived'],df['class'])
```

```python
x
```

```python
#annot (by default None) can be set to True to show values printed in each square
#fmt used to remove roundabout values
#cmap used to change the color of whole graph.
sns.heatmap(x, annot=True, fmt='.1f', cmap='coolwarm')
```

```python
sns.barplot(x='class', y='age', hue='sex', data=df)
#this is also categorical but we had to use one numerical that is "age"
```

```python
sns.stripplot(x='class', y='age', hue='sex', data=df)
```

```python
df.plot(x='class', y='age', kind='scatter')
```

```python
# Boxplot
df.boxplot(column='age', by='class')
```
