```python
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset('titanic')
df.head(2)
```

```python
sns.barplot(data = df, x='class', y='age', hue='class')
```

```python
sns.barplot(data = df, x='class', y='fare', hue='class', estimator=sum)
#here we are using estimator=sum that basically adds up all the numeric values as shown below.
```

```python
sns.barplot(data = df, x='class', y='age', hue='sex')
```

```python
sns.barplot(data = df, x='class', y='fare', hue='sex', estimator=sum)
```

```python
sns.boxplot(data = df, x='class', y='age', hue='class')
```

```python
sns.boxplot(data = df, x='class', y='age', hue='survived')
```

```python
sns.violinplot(x='class', y='age', data=df)
```

```python
sns.distplot(df[df['class'] == 'Second']['age']) #checking the distribution of age of second class specifically.
```
